```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Sample data (can replace with real data)
data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'income (USA)': [770, 780, 790, 795, 791],
    'income (Canada)': [790, 800, 810, 820, 825],
    'income (Germany)': [780, 790, 800, 810, 812]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the 'Year' as the index
df.set_index('Year', inplace=True)

# Plot the data
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, palette="tab10", linewidth=2)

# Add titles and labels
plt.title('income Over Time by Country')
plt.xlabel('income (Years)')
plt.ylabel('Year')

# Show the plot
plt.legend(title='Country')
plt.show()
```
